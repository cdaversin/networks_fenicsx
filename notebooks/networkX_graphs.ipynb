```python
### --- Using networkX and GMSH to display the set of graphs available from mesh_generation.py --- ###
```

```python
from networks_fenicsx.config import Config
from networks_fenicsx.mesh import mesh_generation
import networkx as nx
```

```python
# Setup configuration
cfg = Config()
node_color='#87cefa'
```

```python
# Line graph
G = mesh_generation.make_line_graph(n=3, cfg=cfg, dim=2)

pos = nx.get_node_attributes(G, 'pos')
nx.draw_networkx(G, pos=pos, arrowsize=15, node_color=node_color)
```

```python
# Simple bifurcation
G = mesh_generation.make_Y_bifurcation(cfg=cfg, dim=2)

pos = nx.get_node_attributes(G, 'pos')
nx.draw_networkx(G, pos=pos, arrowsize=15, node_color=node_color)
```

```python
# Double bifurcation
G = mesh_generation.make_double_Y_bifurcation(cfg=cfg, dim=2)

pos = nx.get_node_attributes(G, 'pos')
nx.draw_networkx(G, pos=pos, arrowsize=15, node_color=node_color)
```

```python
# Tree - number of generations n=4
n = 4
G = mesh_generation.make_tree(n=n, H=n, W=n, cfg=cfg, dim=2)

pos = nx.get_node_attributes(G, 'pos')
nx.draw_networkx(G, pos=pos, arrowsize=15, node_color=node_color)
```

```python
# Honeycomb - 3 x 3
n=3
m=3
G = mesh_generation.make_honeycomb(n=n, m=m, cfg=cfg, dim=2)

pos = nx.get_node_attributes(G, 'pos')
nx.draw_networkx(G, pos=pos, arrowsize=15, node_color=node_color)
```
